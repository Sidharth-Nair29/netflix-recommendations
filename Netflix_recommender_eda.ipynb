```python
df=df[['title', 'description', 'type', 'genres', 'cast', 'director']]
```

```python
df=df.fillna('Unknown')
```

```python
df = df.reset_index(drop=True)
df['combined_cols']=(df['description']*2 + ' ' + df['genres']*3 + ' ' + df['cast'] + ' ' + df['director']).str.lower()
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df['combined_cols'])
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print("Cosine similarity matrix shape:", cosine_sim.shape)

indices = pd.Series(df.index, index=df['title'].str.lower()).drop_duplicates()
```

```python
def recommend(title, cosine_sim=cosine_sim):
    title=title.lower()
    if title not in indices:
        return f"❌ '{title}' not found in dataset."

    # Get the index of the given title
    idx = indices[title]

    # Find type of the input title (Movie or TV Show)
    input_type = df.iloc[idx]['type']

    # Get pairwise similarity scores for this title
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort by similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:]

    recommendations = []
    for i, score in sim_scores:
        # Only add items of the same type
        if df['type'].iloc[i] == input_type:
            recommendations.append((df['title'].iloc[i], round(float(score), 2)))
        if len(recommendations) == 5:  # top 5
            break

    return recommendations
```

```python
print(f"Here are recommendations based on your choice of movie : {recommend("blue exorcist")}")
```
